```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\sivan\\Projects\\real-time-fraud-system\\data\\raw\\creditcard.csv")
df.head()
```

```python
df['Class'].value_counts(normalize=True)
```

```python
plt.figure(figsize=(15,15))
sns.histplot(df["Amount"], kde = True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data = df, x = "Amount", y= "Class")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.boxplot(data=df, x="Class", y="Amount", ax=ax)
plt.show()
```

```python
df['Hour'] = (df['Time'] // 3600) % 24
```

```python
import matplotlib.pyplot as plt

transactions_per_hour = df.groupby('Hour').size()

plt.figure(figsize=(10,5))
transactions_per_hour.plot(kind='bar')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Transactions')
plt.title('Transactions per Hour')
plt.show()
```

```python
fraud_rate_per_hour = df.groupby('Hour')['Class'].mean()

plt.figure(figsize=(10,5))
fraud_rate_per_hour.plot(kind='bar')
plt.xlabel('Hour of Day')
plt.ylabel('Fraud Rate')
plt.title('Fraud Rate per Hour')
plt.show()
```
